I told the program that 42, 32, and 27 were all higher than my number (I chose 24). The computer goes on to guess 6, so I respond that its guess is lower than my number. The next guess after that is 41, which is higher than both 27 and 32, which the computer already guessed and I already told it they were higher than my number. So, this version is not keeping a record of older guesses, only the very last guess. This way it could take the computer forever to correctly guess my number. My version fixes this problem by keeping a record of *all* the old guesses. 

**TRY OUT THEIR VERSION NOW**

````python
#   THEIR VERSION, NOT MINE

import random
import numpy as np

computer_tries = 0
player_number  = None
computer_guess = random.randint(1, 100)

print(
    """
    Welcome Player to the fabulous number guessing game.
    Please allow me to show you my incredible deduction skills
    """)

question = None
while question != ("yes"):
    question = input("Has the player picked a number? ")
    #question = question.lower()
    if question == "yes":
        print("\nI will now guess your number!!!\n")
        while player_number == None:
            computer_tries += 1
            print(computer_guess)
            confirmation = input("Is this the correct number? ")
            #confirmation = confirmation.lower()
            if confirmation == "yes":
                player_number = computer_guess
                if computer_tries < 2:
                    print("I did it! I guessed your number was", computer_guess,
                           "and it only took me", computer_tries,
                           "try to get it right!")
                else:
                    print("I did it! I guessed your number was", computer_guess,
                           "and it only took me", computer_tries,
                           "tries to get it right!")
            else:
                higher_lower = None
                while higher_lower not in ("higher", "lower"):
                    higher_lower = input("Is my guess higher or lower"
                                + " than your number? ")
                    higher_lower = higher_lower.lower()
                    if higher_lower == "lower":
                        higher = computer_guess
                        computer_guess = random.randint(higher, 101)
                    elif higher_lower == "higher":
                        lower = computer_guess
                        computer_guess = random.randint(0, lower)
                    else:
                        print("Please choose either higher or lower.")
````

## My version of the Game.
I set this up as a function so it can be used multiple times. 

In this notebook, I chose the number __24__ every time I ran my function. 

__Make sure you cleared the notebook session if you ran their version (above) before running mine.__

````python
import random
import numpy as np

def guess_my_number():
    
    print(
        """

        Pick a number between 1 and 100 and I will guess it. 

        """)
    
    
    termination_message1 = '''
    ************************************************

    Clearly you do not want to cooperate. 
    ENDING GAME !!

    ************************************************
    '''
    
    termination_message2 = '''
    ************************************************

    You either messed up or you are lieing!
    ENDING GAME !!

    ************************************************
    '''
        

    nGuesses = 1 # first guess counts as a try.
    number = None   # your number 
    guess = random.randint(1, 100)  # random guess from 1 to 100.

    # we will store all guesses in a dictionary of keys 0 and 1.
    # numbers too high will be stored in key 1 and numbers too 
    # low will be stored
    # in key 0. This is so we narrow down our guesses as we go. 
    guesses    = dict()
    guesses[0] = []
    guesses[1] = []    
    
    yes_options = ("y","ye","yes","yep", "ya", "yea", "yeah")
    no_options  = ("n","no","na","nah", "nop", "nope")
    
    question = 'no'  # use to get while loop started.
    retries_main = 4  # retries for main response.
    
    while question.lower() not in yes_options:
        question = input("\n\nIs your number %i ? \n"%guess)
        question = question.replace(' ','') # remove whitespaces
        
        if question.lower() in yes_options:
            guessed_message = '''
                Great News!
                I guessed your number and it took me %i %s!!!
                '''%(nGuesses, np.where(nGuesses < 2,'try','tries'))
            print(guessed_message)
            break
        elif question.lower() in no_options:
            # this is what we want
            # set these up for rest of program
            lower_higher = None # needs to be False or None for while statement to run.
            retries = 4         # 4 retries before program fails. 
            lower_higher_message = '''
                My guess was %i
                
                Press 0 if I was too low and 1 if I was too high.

                    0 - too low
                    1 - too high

                '''%guess
            pass

            while (lower_higher not in ('0','1')) and (retries > 0):
                # if retries < 1, force this while loop to end and
                # enter program termination at next if statement.
                lower_higher = str(input(lower_higher_message))
                lower_higher = lower_higher.replace(' ','') # remove whitespaces

                if lower_higher == '0':
                    # guesses that were too low
                    guesses[0].append(int(guess))

                elif lower_higher == '1':
                    # guesses that were too high
                    guesses[1].append(int(guess))
                else:
                    print("\n \n  *** PLEASE ONLY USE 0 or 1 ***  \n")
                    retries = retries - 1

            if retries < 1:
                print(termination_message1)
                break
            else:
                pass
            
            lower_higher_response = '''
            Okay, %i is too %s.
            '''%(guess, np.where(lower_higher=='0', 'low','high'))
            print( lower_higher_response )
            
                  
            try:
                # update lower boundary to be max of all other guesses
                lower_bound = max(guesses[0]) 
            except:
                lower_bound = 1
            try:
                upper_bound = min(guesses[1]) # upper boundary
            except:
                upper_bound = 100

            try:
                # we want +1 and -1 to exclude those already guessed.
                guess = random.randint(lower_bound+1, upper_bound-1) 
            except:
                try:
                    #raise Exception("You must be a liar ;)")
                    break
                finally:
                    print(termination_message2)
            nGuesses = nGuesses + 1
        
        else:  
            # THIS BELONGS TO THE FIRST WHILE LOOP'S IF STATEMENTS.
            # IF USER DOESN'T USE ANY OF THE yes_options or no_options,
            # LET THEM RETRY THEIR RESPONSE.
            #print("\n \n  *** PLEASE ONLY USE y or n ***  \n")
            print("\n \n *** PLEASE ONLY USE ON OF THE FOLLOWING RESPONSES: ***")
            print("For yes:  %r \nFor no:   %r \n\n"%(yes_options, no_options))
            retries_main = retries_main-1
        
        if retries_main < 1:
                print(termination_message1)
                break
        else:
            pass
            
````

````python
guess_my_number()
````
